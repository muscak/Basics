##Indexing, Selection and Slicing
Each columns in the DataFrame can be selected with regular indexing. 
Each column is represented as a `Series`.

```python
df
```

```python
# Select individual row by index. Returns Pandas Series.
df.loc['Canada']
```

```python
# Select individual row by sequential position. Returns Pandas Series.
df.iloc[-1]
```

```python
#Select a specific column. Returns Pandas Series.
df['Population']
```

```python
# Get the selected elements in DataFrame type.
df['Population'].to_frame()
```

```python
#Select specific columns using a list. Returns Pandas Series.
columns = ['Population', 'HDI']
df[columns]
```

```python
# Combined selection using indexing
df.loc['France':'Italy', ['Population', 'GDP']]
```

```python
# Combined selection using sequential position
df.iloc[1:3, [0, 3]]
#df.iloc[1:3, 1:3] another method
```

###Conditional Selection

```python
df
```

```python
# Creating a 2 dimentional DataFrame. X axis of the plot must be the index of the DataFrame. 
# So, the Year column will be the index. If I filter only by Country I'll get same value over and over again for each Energy_type.
# Since .loc filtering is immutable, I need to craete another DataFrame in the memmory to be used later.
df_USA = df.loc[(df['Country'] == 'United States') & (df['Energy_type'] == 'all_energy_types'), ['Year', 'CO2_emission']]
# Setting the Year as the index of the new DataFrame.
df_USA = df_USA.set_index('Year')
# Plotting using the pandas integration with Matplotlib
df_USA.plot()
```

```python
# Plotting with matplotlib.pyplot interface
plt.plot(df_USA.index, df_USA['CO2_emission'])
```

```python
x = np.arange(-10,11)
```

```python
# Each plot fonction alters the global state. 
# If you want to set the settings of your plot, you can use the plt.figure function. plt.title keeps altering the global plot
plt.figure(figsize=(12, 6))
plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
plt.title('Nice Plot')
```
