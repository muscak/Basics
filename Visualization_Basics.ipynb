```python
x = np.arange(-10, 11)
```

##Global API
Matplotlib's default pyplot API has a global, MATLAB-style interface.

```python
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')
plt.plot(x, x ** 2)
plt.plot(x, -1 * (x ** 2))
```

```python
# Visualization of 2 different plot in the same figure
plt.figure(figsize=(12, 6))
plt.title('My Nice Plot')

plt.subplot(1, 2, 1) # rows, colums, panel selected (every action after this line will be impacting the selected plot)
plt.plot(x, x ** 2)
plt.plot([0, 0, 0], [-10, 0, 100])
plt.legend(['X^2', 'Vertical Line'])
plt.xlabel('X')
plt.ylabel('X Squared')

plt.subplot(1, 2, 2)
plt.plot(x, -1 * (x ** 2))
plt.plot([-10, 0, 10], [-50, -50, -50])
plt.legend(['-X^2', 'Horizontal Line'])
plt.xlabel('X')
plt.ylabel('X Squared')
```

##Object Oriented API

```python
# Creating figures and axes in object oriented aproach. 
fig, axes = plt.subplots(figsize=(12, 6))
```

```python
axes.plot(
    x, (x ** 2), color='red', linewidth=3,
    marker='o', markersize=8, label='X^2'
)

axes.plot(x, -1 * (x ** 2), 'b--', label='-X^2') # 'b--' means the color of the plot is blue and in dashed style

# Since we created an object, we need call the appropriote functions to set the properties of the plot
axes.set_xlabel('X') # Label the X axis
axes.set_ylabel('X Squared') # Label the Y axis

axes.set_title('My Nice Plot') # Set the title

axes.legend() # Display the plot labels on the figure

fig # Show the figure object
```

```python
# An example for linestyle
fig, axes = plt.subplots(figsize=(12, 6))

axes.plot(x, x + 0, linestyle='solid')
axes.plot(x, x + 1, linestyle='dashed')
axes.plot(x, x + 2, linestyle='dashdot')
axes.plot(x, x + 3, linestyle='dotted')

axes.set_title('My Nice Plot')
```

```python
# Another example for plot style. marker, color, linestyle and linewidth keywords can be used too.
fig, ax = plt.subplots(figsize=(12, 6))

ax.plot(x, x + 0, '-og', label='solid green') # solid line with dot markers in green
ax.plot(x, x + 1, '--c', label='dashed cyan') # dashed line in cyan
ax.plot(x, x + 2, '-.b', label='dashdot blue') # dashed and dotted line in blue
ax.plot(x, x + 3, ':r', label='dotted red') # dotted line in red

ax.set_title('My Nice Plot')
ax.legend()
```

```python
# List of the markers
print('Markers: {}'.format([m for m in plt.Line2D.markers]))
```

```python
linestyles = ['_', '-', '--', ':']

print('Line styles: {}'.format(linestyles))
```

##Other Types of Plots

###Figures and Subfigures

```python
# Create single figure with 2x2=4 axis
plot_objects = plt.subplots(nrows=2, ncols=2, figsize=(14, 6))

fig, ((ax1, ax2), (ax3, ax4)) = plot_objects
plot_objects
```

```python
ax1.plot(np.random.randn(50), c='red', linestyle='--')
ax2.plot(np.random.randn(50), c='green', linestyle=':')
ax3.plot(np.random.randn(50), c='blue', marker='o', linewidth=3.0)
ax4.plot(np.random.randn(50), c='yellow')

fig
```

```python
# More complex example
plt.figure(figsize=(14, 6))

ax1 = plt.subplot2grid((3,3), (0, 0), colspan=3)  # Merges 3 columns. Positions to the 1st row and 1st column of the axes array.
ax1.set_title('Axis 1', y=0.5)                    # 'y' sets the vertical position of the title. Default value is 1.0 and it means the top of the axes. 
ax2 = plt.subplot2grid((3,3), (1, 0), colspan=2)  # Merges 2 columns. Positions to the 2nd row and 1st column of the axes array.
ax2.set_title('Axis 2', y=0.5)
ax3 = plt.subplot2grid((3,3), (1, 2), rowspan=2)  # Merges 2 rows. Positons to the 2nd row and 3rd column
ax3.set_title('Axis 3', y=0.5)
ax4 = plt.subplot2grid((3,3), (2, 0))
ax4.set_title('Axis 4', y=0.5)
ax5 = plt.subplot2grid((3,3), (2, 1))
ax5.set_title('Axis 5', y=0.5)
```

###Scatter Plot

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)
colors = np.random.rand(N)
area = np.pi * (20 * np.random.rand(N))**2 # 0 to 15 point radii
```

```python
# Encoding 4 dimensions in one plot: x, y, size of the buble and color of the buble.
plt.figure(figsize=(14, 6))
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Spectral') # Set the transparency to 50%
plt.colorbar() # Display the color bar.

plt.show()
```

```python
fig = plt.figure(figsize=(14, 6))

ax1 = fig.add_subplot(1, 2, 1)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel1')
plt.colorbar()

ax2 = fig.add_subplot(1, 2, 2)
plt.scatter(x, y, s=area, c=colors, alpha=0.5, cmap='Pastel2')
plt.colorbar()

plt.show()
```

For more `cmap` options visit [here](https://matplotlib.org/stable/api/matplotlib_configuration_api.html?highlight=colormaps#matplotlib.colormaps).

###Histograms

```python
values = np.random.rand(1000)
```

```python
plt.subplots(figsize=(12, 6))

# Group the values in the size of 10 bins. Use steel blue color with green border for bars. 
plt.hist(values, bins=10, alpha=0.8,
         histtype='bar', color='steelblue',
         edgecolor='green')

# The limit of the x axis ticks is between -5 to 5
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

```python
# Saving a plot
fig.savefig('hist.png')
```

###KDE (Kernel Density Estimation)

```python
from scipy import stats

density = stats.kde.gaussian_kde(values)
density
```

```python
plt.subplots(figsize=(12, 6))

values2 = np.linspace(min(values)-10, max(values)+10, 100)

plt.plot(values2, density(values2), color='#FF7F00')
plt.fill_between(values2, 0, density(values2), alpha=0.5, color='#FF7F00')
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

```python
# Combining 2 plots
plt.subplots(figsize=(12, 6))

plt.hist(values, bins=10, alpha=0.8, density=1, # density to bring the max values of 'values' to the same as values2 on the y axis
         histtype='bar', color='steelblue',
         edgecolor='green')
plt.plot(values2, density(values2), color='#FF7F00', linewidth=3)
plt.xlim(xmin=-5, xmax=5)

plt.show()
```

###Bar Plots

```python
Y = np.random.rand(1, 5)[0]
Y2 = np.random.rand(1, 5)[0]
```

```python
plt.figure(figsize=(12, 4))

bar_width = 0.5
plt.bar(np.arange(len(Y)), Y, width=bar_width, color='#00b894')
plt.show()
```

```python
# Stacked bar plot

plt.figure(figsize=(12, 4))

bar_width = 0.5
plt.bar(np.arange(len(Y)), Y, width=bar_width, color='#00b894', label='Label Y')
plt.bar(np.arange(len(Y2)), Y2, width=bar_width, color='#e17055', bottom=Y, label='Label Y2')
plt.legend()
plt.show()
```
